```python
import torch
```

```python
x = torch.arange(4.0)
x
```

```python
x.requires_grad_(True)
x.grad
```

```python
y = 2 * torch.dot(x, x)
y
```

```python
y.backward()
x.grad
```

```python
x.grad == 4 * x
```

```python
x.grad.zero_()
y = x.sum()
y.backward()
x.grad
```

```python
x.grad.zero_()
y = x * x
u = y.detach()
z = u * x

z.sum().backward()
x.grad == u
```

```python
x.grad.zero_()
y.sum().backward()
x.grad == 2 * x
```

```python
def f(a):
    b = a * 2 
    while b.norm() < 1000:
        b = b * 2
    if b.sum() > 0:
        c = b 
    else: 
        c = 100 * b
    return c 
```

```python
a = torch.randn(size=(), requires_grad=True)
d = f(a)
d.backward()
```

```python
a.grad == d / a 
```
